Number Guessing Game in Python

This is a simple game where the computer generates a random number and the player tries to guess it. 
The program gives hints such as "Too High" or "Too Low" until the correct number is guessed.

```python
import random
```

```python
number=random.randint(1,100)
```

```python
import random

play='yes'

while play=='yes':
    print("\n==============================")
    print("\n🎮 Welcome to Number Guessing Game!")
    print("==============================\n")
    print("----------- GAME MENU -----------")
    print("Choose difficulty level")
    print("1. Easy (1-50)")
    print("2. Medium (1-100)")
    print("3. Hard (1-200)")
    print("==============================\n")

    level=int(input("Enter choice: "))

    if level==1:
        number=random.randint(1,10)
    elif level==2:
        number=random.randint(1,100)
    elif level==3:
        number=random.randint(1,200)
    else:
        print("Invalid choice")
        continue

    max_attempts=5
    attempts=0

    while attempts < max_attempts:

        guess=int(input("Enter your guess: "))
        attempts+=1

        print("Attempt",attempts, "of", max_attempts)

        if guess==number:
            print("\n---------GAME RESULT---------")
            print("🎯 Perfect guess!")
            print("🎉 Correct! You guessed the number!")
            print("------------------------------\n")
            break

        elif guess>number:
            print("📈 Too high!")

        else:
            print("📉 Too low!")

        if abs(guess-number)==1:
            print("🔥 You're very close!")

    if guess!=number:
        print("💀 Game Over!")
        print("The number was:", number)
        print("\n------------------------------")
    play=input("Do you want to play again? (yes/no): ")
    print("------------------------------\n")

print("Thanks for playing! 👋")
```
